```python
import numpy as np
import pandas as pd

np.random.seed(0)
num_data = 100
data = np.exp(np.random.uniform(size = num_data) * 4)
df = pd.DataFrame(data.T, columns = ["data"])
```

```python
print("The 75th percentile is:", df.quantile(q = 0.75))
df.plot.hist()
```

```python
df.plot.box()
```

```python
# The histogram does a good job at showing that there is a larger frequency of low cost or inexpensive drugs as compared to high cost drugs. The right skew of this graph indicates this to be the case. 
# It is not great at telling you the important values in the data like mean and median or quartiles. I would add graphics within the graph to show these metrics as well as clarify the x axis and be more clear what it is showing. 

# The box plot has something of the opposite problem, it shows the important values as mean, median, and quartiles, but is a less used and less intuitive graph. It also highlights the outliers in the expensive drug category much more, which for the intended analysis is better
# I would improve this graph by marking the important metrics more clearly and adding a y axis label for the data to better clarify for readers. 
# 
# Overall I would choose the boxplot as the objective is to show the relative price of the drugs and this plot does much better as it includes mean and quartiles which provide greater context for the outliers, as the intended focus. 
```
